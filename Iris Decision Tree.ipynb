```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
df = pd.read_csv('C:/Users/KEMI- ITB CONCRETE/Downloads/datasetsIris.csv')
df
```

```python
#Check data summary
df.describe()
```

```python
#Check relationship between columns
sns.pairplot(df,hue='Species')
```

```python
#Split Dataset into train and test
df_inputs = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
df_inputs
```

```python
df_classes = df['Species'].values
df_classes
```

```python
#Train_Test_Split
(train_inputs,test_inputs,train_classes,test_classes) = train_test_split(df_inputs,df_classes,train_size=0.7,random_state=1)
(train_inputs,test_inputs,train_classes,test_classes)
```

```python
(train_inputs,test_inputs,train_classes,test_classes) = train_test_split(df_inputs,df_classes,test_size=0.3,random_state=1)
(train_inputs,test_inputs,train_classes,test_classes)
```

```python
#Decision Tress
dtc = DecisionTreeClassifier()
dtc.fit(train_inputs,train_classes)
dtc.score(test_inputs,test_classes)
```

```python
from sklearn import tree
text_representation = tree.export_text(dtc)
print(text_representation)
```

```python
with open("decistion_tree.log", "w") as fout:
    fout.write(text_representation)
```
